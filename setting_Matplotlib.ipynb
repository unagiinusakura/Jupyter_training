## 5-1 さまざまな図形の描画

### 円弧を描画する
- matplotlib.patches.Arcで描画
- 引数xyにタプル型で中心座標
- 引数widthとheightに幅と高さ
- angleに度数で回転角度
- theta1とtheta2は描画開始と終了の角度を指定(角度は時計の3時を基準)

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline

# Arc円弧
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

c1 = mpatches.Arc(xy=(0.4, 0.5), width=0.4, height=0.4, angle=30,
                 theta1=0, theta2=230, linewidth=10, color='lightblue')

c2 = mpatches.Arc(xy=(0.6, 0.5), width=0.4, height=0.4, angle=30,
                 theta1=0, theta2=230, linewidth=10, color='pink')

ax.add_patch(c1)
ax.add_patch(c2)

plt.show()
```

### 矢印を描画する
- mpatches.Arrowで描画
- 引数xyに始点のX座標とY座標
- dxとdyで始点の座標からの幅と高さ
- widthで矢印の幅

```python
# Arrow矢印
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

a1 = mpatches.Arrow(x=0.1, y= 0.4, dx=0.6, dy = 0.5, width=0.5, color='lightblue')
a2 = mpatches.Arrow(x=0.3, y=0.1, dx=0.6, dy = 0.5, width=0.5, color='pink')

ax.add_patch(a1)
ax.add_patch(a2)

plt.show()
```

### 矢印の書式をより詳細に設定したい場合
- matplotlib.patches.FancyArrowで描画
- 引数head_widthは矢印先端の幅
- head_lengthは矢印先端の長さ
- shapeは矢印の形状指定で,full(全表示),left(左側表示),right(右側表示)の指定可能

```python
# Fancy Arrow矢印
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

a1 = mpatches.FancyArrow(x=0.2, y=0.3, dx=0.4, dy=0.3, width=0.1,
                        head_width=0.3, head_length=0.2, shape='right', color='lightblue')

a2 = mpatches.FancyArrow(x=0.2, y=0.3, dx=0.4, dy=0.3, width=0.1,
                        head_width=0.3, head_length=0.2, shape='left', color='pink')

ax.add_patch(a1)
ax.add_patch(a2)

plt.show()
```

### 円形を描画する
- matplotlib.patches.Circleで描画
- 引数xyにタプル型で中心座標
- radiusで半径を指定

```python
# Circle 円形
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

c1 = mpatches.Circle(xy= (0.5, 0.5), radius=0.4, color='lightblue')
c2 = mpatches.Circle(xy= (0.5, 0.5), radius=0.2, color='white')
ax.add_patch(c1)
ax.add_patch(c2)

plt.show()
```

### 正多角形を描画する
- matplotlib.patches.CirclePolygonで描画
- resolutionで円を近似する際の返の数

```python
# Circle Polygon 正多角形
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

cp1 = mpatches.CirclePolygon(xy= (0.5, 0.5), radius=0.4, resolution=6, color='lightblue')
cp2 = mpatches.CirclePolygon(xy= (0.5, 0.5), radius=0.2, resolution=8, color='white')

ax.add_patch(cp1)
ax.add_patch(cp2)

plt.show()
```

### 楕円形を描画する
- matplotlib.patches.Ellipseで描画
- 引数xyで中心座標
- widthとheightで幅と高さ
- angleで回転角度

```python
# Ellipse 楕円形
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

e1 = mpatches.Ellipse(xy=(0.5, 0.5), width=0.8, height=0.5, angle=30, color='lightblue')
e2 = mpatches.Ellipse(xy=(0.5, 0.5), width=0.8, height=0.2, angle=30, color='white')

ax.add_patch(e1)
ax.add_patch(e2)

plt.show()
```

### 扇形を描画する
- matplotlib.patches.Wedgeで描画
- 引数centerで中心座標、rで半径
- theta1,theta2で切れ込み(くさび)の始点と終点の角度

```python
# Wedge 扇形
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

w1 = mpatches.Wedge(center=(0.3,0.55), r= 0.25, theta1=30, theta2=-30, color='pink')
w2 = mpatches.Wedge(center=(0.7,0.4), r= 0.25, theta1=210, theta2=150, color='lightblue')

ax.add_patch(w1)
ax.add_patch(w2)

plt.show()
```

### 長方形を描画する
- matplotlib.patches.Rectangleで描画
- 引数xyで長方形左下角の座標
- widthとheightで指定した座標からの幅と高さ
- angleで回転角度

```python
# Rectangle 長方形
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

r1 = mpatches.Rectangle(xy=(0.1, 0.4), width=0.5, height=0.5, angle=0, color='lightblue')
r2 = mpatches.Rectangle(xy=(0.4, 0.1), width=0.5, height=0.5, angle=0, color='pink')

ax.add_patch(r1)
ax.add_patch(r2)

plt.show()
```

### 多角形を描画する
- matplotlib.patches.Polygonで描画
- 引数xyで多角形の各頂点の座標を指定

```python
# Polygon 多角形
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

p1 = mpatches.Polygon(xy=[(0,0.2), (0.3, 0.8), (0.6, 0.2)], facecolor='lightblue')
p2 = mpatches.Polygon(xy=[(0.4,0.8), (0.7, 0.2), (1.0, 0.8)], facecolor='pink')

ax.add_patch(p1)
ax.add_patch(p2)

plt.show()
```

## 5-3 色の指定とカラーマップ

#### 色と不透明度を設定する
- color : オブジェクトの色
- facecolor : 塗りつぶし色
- edgecolor : 枠線の色

- 凡例の不透明度の設定 : framealpha

```python
x = np.arange(0, 6, 0.1)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure()
ax = fig.add_subplot(111)

# 描画
ax.scatter(x,y, label='sin', color='lightblue', edgecolor='blue')
ax.scatter(x,z, label='cos', color='lightblue', edgecolor='blue', alpha=0.3)

ax.legend(framealpha=1, loc='lower right')

plt.show()
```

### 色を指定する
- Matplotlibカラーストリング: 'b','g'
- RGBカラーコード: (70,130,180)
- RGBaカラーコード: (255,215,0,1)
- 16進数カラーコード: '#4682b4','#ffd700'
- グレースケール: 0.75
- HTMLカラーネーム: 'red','blue'
- X11/CSS4カラーネーム: 'Red','Blue'
- xkcdカラー調査ネーム: 'xkcd:sky blue'
- 正規表現カラー: 'C0', 'C1'

```python
#色指定の例
fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot(111, xlim=(0,14))

x = np.arange(0,3, 0.1)
y = np.cos(x)

# 各種方法による色指定で描画
ax.scatter(x, y, label='color_string', color='b')
ax.scatter(x+ 0.5, y, label='RGBA tuple', color=(0.1,0,0.5,1))
ax.scatter(x+ 1.0, y, label='RGBA 16', color='#330099')
ax.scatter(x+ 1.5, y, label='Gray Scale', color='0.3')
ax.scatter(x+ 2.0, y, label='HTML color', color='DarkOliveGreen')
ax.scatter(x+ 2.5, y, label='X11/CSS', color='DarkSeaGreen')
ax.scatter(x+ 3.0, y, label='Tegular expression', color='C8')
ax.scatter(x+ 3.5, y, label='Tableau', color='tab:gray')

ax.legend()
plt.show()
```

#### カラーマップを利用する
- matplotlib.cmモジュールを用いて設定

```python
import matplotlib.cm as cm

fig = plt.figure()
ax = fig.add_subplot(111)

x = np.arange(20)
y = np.arange(20)

ax.scatter(x, y +5, color=cm.spring(0.5), label='spring(0.5)')
ax.scatter(x, y, color=cm.get_cmap('winter')(0.5), label='winter(0.5)')

ax.legend()
plt.show()
```

#### 段階的に色を変化させる場合
- 引数cmapにカラーマップを指定する

```python
fig = plt.figure(figsize=(4.5,2))
ax = fig.add_subplot(111)

n = 20
x = np.arange(n)
y = np.ones(n)

ax.scatter(x, y, c=x, cmap= cm.Set1)

plt.show()
```

```python
# 個別に色を指定した散布図
fig = plt.figure(figsize=(4.5, 2))
ax = fig.add_subplot(111)

n = 20
x = np.arange(n)
y = np.ones(n)

for i in range(n):
    color = cm.Set1(i / n * 0.5)
    ax.scatter(x[i], y[i], c=color)
    
plt.show()    
```

## 5-4 線の書式設定

#### 線に設定できる書式の例
- color :線の色
- linewith :線の太さ 
- linestyle :線の種類 
- solid_capstyle :線の先端の形状 
- solid_joinstyle :線の結合点の形状 
- dash_capstyle :破線・点線の先端の形状
- dash_joinstyle :破線・点線の結合点の形状

### 線の太さを設定する

```python
x = np.array([0,1])
y = [0,1]

fig = plt.figure(figsize=(9,3))
ax = fig.add_subplot(111, xlim=(-0.2, 4))

# 線の太さと不透明度を設定して描画
ax.plot(x, y, linewidth=1, color='b', label='linewidt=1(alpha=1.0)')
ax.plot(x + 0.1, y, linewidth=1, color='b',alpha=0.5, label='linewidt=1(alpha=0.5)')

ax.plot(x + 0.3, y, linewidth=2, color='b', label='linewidt=2(alpha=1.0)')
ax.plot(x + 0.4, y, linewidth=2, color='b',alpha=0.5, label='linewidt=2(alpha=0.5)')

ax.plot(x + 0.7, y, linewidth=5, color='b', label='linewidt=5(alpha=1.0)')
ax.plot(x + 0.7, y, linewidth=5, color='b',alpha=0.5, label='linewidt=5(alpha=0.5)')

ax.plot(x + 1.1, y, linewidth=10, color='b', label='linewidt=10(alpha=1.0)')
ax.plot(x + 1.3, y, linewidth=10, color='b',alpha=0.5, label='linewidt=10(alpha=0.5)')

ax.legend()

plt.show()
```

### 線の先端の形状を設定する

#### 線の先端の形状を設定
1. 線の先端の形状はsolid_capstyleで設定
- butt: 線の進行方向に対して垂直にフラットな形状(デフォルト)
- round: 線の端に半円が追加された形状
- projecting: 線の端の半分の長方形が追加された形状

```python
x = [0.1,1]
y = np.array([0,0])

fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot(111, xlim=(0, 1.7), ylim=(-0.2, 1.2))

# 線の先端形状を設定して描画
ax.plot(x, y+1, linewidth=12, solid_capstyle='butt', label='butt')
ax.plot(x, y+0.5, linewidth=12, solid_capstyle='round', label='round')
ax.plot(x, y, linewidth=12, solid_capstyle='projecting', label='projecting')

ax.legend(loc=1)
plt.show()
```

### 線の結合点の形状を設定する
1. 線の結合点の形状はsolid_joinstyleで設定
- miter: 留め継ぎ
- round: 円形
- bevel: 面取り

```python
x = np.array([0, 1, 2, 3])
y = [0, 1.0, 0.2, 1.2]

fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot(111, xlim=(0, 4.5))

#線の結合点の形状を設定して描画
ax.plot(x, y, linewidth=10, solid_joinstyle='miter', label='miter')
ax.plot(x + 0.5, y, linewidth=10, solid_joinstyle='round', label='round')
ax.plot(x + 1.0, y, linewidth=10, solid_joinstyle='bevel', label='bevel')

ax.legend()
plt.show()
```

### 線の種類を設定する
1. 線の種類はlinestyleで設定する
- 文字列: 'solid','dashed'
- 記号: '-','--'
- タプル: (offset, (on, off))

```python
x = np.array([0, 1])
y = np.array([1.9, 1.9])

fig = plt.figure(figsize=(6,4))
ax = fig.add_subplot(111, xlim=(-0.1, 1.7), ylim=(0, 2))

# 線種を設定して描画
ax.plot(x, y, linewidth=5, color='b', linestyle='solid', label='solid')
ax.plot(x, y -0.1, linewidth=5, color='b',alpha=0.5, linestyle='-', label='-')

ax.plot(x, y -0.4, linewidth=5, color='b', linestyle='dashed', label='dashed')
ax.plot(x, y -0.6, linewidth=5, color='b',alpha=0.5, linestyle='--', label='--')

ax.plot(x, y -0.9, linewidth=5, color='b', linestyle='dashdot', label='dashdot')
ax.plot(x, y -1.0, linewidth=5, color='b',alpha=0.5, linestyle='-.', label='-.')

ax.plot(x, y -1.3, linewidth=5, color='b', linestyle='dotted', label='dotted')
ax.plot(x, y -1.4, linewidth=5, color='b',alpha=0.5, linestyle=':', label=':')

ax.plot(x, y -1.7, linewidth=5, color='g', linestyle=(0,(5,2)), label='(0, (5,2))')
ax.plot(x, y -1.8, linewidth=5, color='g', linestyle=(4,(5,2)), label='(4, (5,2))')

ax.legend()
plt.show()
```

### 破線・点線の先端の形状を設定する
- 破線・点線の先端の形状はdash_capstyleで指定

```python
x = [0.1, 1]
y = np.array([0,0])

fig = plt.figure()
ax = fig.add_subplot(111, xlim=(0, 1.7), ylim=(-0.2, 1.2))

# 破線・点線の先端形状を設定して描画
ax.plot(x, y+1, linewidth=12, linestyle='-.', dash_capstyle='butt', label='butt')
ax.plot(x, y + 0.5, linewidth=12, linestyle='-.', dash_capstyle='round', label='round')
ax.plot(x, y, linewidth=12, linestyle='-.', dash_capstyle='projecting', label='projecting')

ax.legend()
plt.show()
```

#### 破線・点線の結合点の形状を設定する
- 破線・点線の結合点の形状はdash_joinstyleで指定

```python
x = np.array([0, 1, 2, 3])
y = [0, 1.0, 0.2, 1.2]

fig = plt.figure(figsize=(6,3))
ax = fig.add_subplot(111, xlim=(0, 4.5))

# 破線の結合形状を変えて描画
ax.plot(x, y, linewidth=10, linestyle=(0, (3, 0.5)),dash_joinstyle='miter', label='miter')
ax.plot(x + 0.5, y, linewidth=10, linestyle=(0, (3, 0.5)),dash_joinstyle='round', label='round')
ax.plot(x + 1.0, y, linewidth=10, linestyle=(0, (3, 0.5)),dash_joinstyle='bevel', label='bevel')

ax.legend()

plt.show()
```

## 5-5 フォントとテキストボックスの書式設定

#### フォントの書式
- family: フォントファミリーの種類
- style: フォントスタイル
- weight: フォントの太さ
- size: フォントサイズ
- alpha: 不透明度
- color: フォント色
- bbox: テキストボックスの書式設定
- verticalalignment: 垂直方向の配置
- horizontalalignment: 水平方向の配置
- multialignment: 複数行の配置
- linespacing: 行間
- fontproperties: フォント書式
- rotaion: 回転
- rotation_mode: 回転方法
- backgroundcolor: 背景色

### テキストを描画する
- テキストは描画したい文字列と文字列左下角の座標を指定することで描画できる

```python
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

# テキストの描画
ax.text(2,4, 'Text', size=40)

ax.set_xticks([i * 2 for i in range(6)])
ax.set_yticks([i * 2 for i in range(6)])

# 目盛り線描画
ax.grid(linestyle='-')

plt.show()
```

### フォントの書式を設定する

```python
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

# フォント設定してテキストを描画
ax.text(0.2, 0.6, 'Text', alpha=0.5, color='Red', size=40)

# フォント設定してテキスト描画(辞書使用)
fps = {
    'family': 'fantasy',
    'style': 'italic',
    'weight': 'heavy',
    'size': 40,
    'color': 'Green'
}

ax.text(0.2, 0.3, 'Text', alpha=0.5, color='Red', fontdict=fps)

ticks= [0.2 * i for i in range(6)]
ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.grid(linestyle='-')

plt.show()
```

### テキストボックスの書式を設定する
- テキストボックスの書式は引数bboxで設定

```python
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

# テキストボックスの書式辞書
boxprops = {
    'facecolor': 'pink',
    'edgecolor': 'red',
    'alpha': 0.5,
    'boxstyle': 'Sawtooth',
    'linewidth': 2
}

# テキストボックスの描画
ax.text(0.07, 0.5, 'テキストボックス\nの書式による\n設定', bbox=boxprops, size=18)

ax.text(0.07, 0.2, '塗りつぶし色の設定', bbox={'alpha': 0.2},
       backgroundcolor='Blue', size=16)

ticks = [0.2 * i for i in range(6)]
ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.grid(linestyle='-')
plt.show()
```

### テキストボックスの水平方向の配置を設定する
- 水平方向の配置はhorizontalalignmentで設定

```python
fig = plt.figure(figsize=(3,3))
ax = fig.add_subplot(111)

x, y = [0.5, 0.5], [0, 4]
ax.plot(x,y)

# テキストボックスの描画(水平方向の配置)
ax.text(0.5, 3, 'left', horizontalalignment='left', size=20)
ax.text(0.5, 2, 'center', horizontalalignment='center', size=20)
ax.text(0.5, 1, 'right', horizontalalignment='right', size=20)

ax.set_yticks([i for i in range(5)])
ax.set_xticks([0.5 * i for i in range(3)])

plt.show()
```

### テキストボックスの垂直方向の配置を設定する
- 垂直方向の配置はverticalalignmentで設定

```python
fig = plt.figure(figsize=(8,3))
ax = fig.add_subplot(111)

# ベースライン描画
x, y = [0, 9], [0.5, 0.5]
ax.plot(x, y)

# テキストボックスの描画(垂直方向の配置)
ax.text(0, 0.5, 'bottom', verticalalignment='bottom', size=20)
ax.text(2.5, 0.5, 'baseline', verticalalignment='baseline', size=20)
ax.text(5.5, 0.5, 'center', verticalalignment='center', size=20)
ax.text(8, 0.5, 'top', verticalalignment='top', size=20)

ax.set_xticks([i * 3 for i in range(4)])
ax.set_yticks([0.5 * i for i in range(3)])

plt.show()
```

### テキストボックス内の文字列の水平方向の配置を設定する
- 水平方向の配置はmultialignmentで設定

### テキストボックスを回転する
- テキストボックスの回転はrotationとrotation_modeで設定

```python
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

# ベースライン描画
axes = [ax1, ax2, ax3, ax4]
x, y = [0.5, 0.5], [0, 1]

for ax in axes:
    ax.plot(x, y, color='SteelBlue')
    ax.plot(y, x, color= 'SteelBlue')
    
# サブプロット1～3の描画
ax1.text(0.5, 0.5, 'horizontal', rotation='horizontal', size=12)
ax2.text(0.5, 0.5, 'vertical', rotation= 'vertical', size=12)
ax3.text(0.5, 0.5, 'rotation=45\nwith anchor\nsetting', rotation=45, rotation_mode='anchor', size=12)

# サブプロット4の描画
# 配置比較用テキストボックス描画
boxprops = {
    'facecolor': 'Pink',
    'edgecolor': 'Pink',
    'alpha': 0.5
}
ax4.text(0.5, 0.5, 'rotation=45\nwithout anchor\nsetting', rotation=0,
        bbox=boxprops, color='white', size=12)

# テキスト描画
ax1.text(0, 0.9, '1)', size=16)
ax2.text(0, 0.9, '2)', size=16)
ax3.text(0, 0.9, '3)', size=16)
ax4.text(0, 0.9, '4)', size=16)

# 回転したテキストボックス描画
boxprops = {
    'facecolor' : 'None',
    'edgecolor' : 'Pink'
}
ax4.text(0.5, 0.5, 'rotation=45\nwithout anchor\nsetting',
        bbox=boxprops, rotation=45, size=12)

for ax in axes:
    ax.set_xticks([0.5 * i for i in range(3)])
    ax.set_yticks([0.5 * i for i in range(3)])
    
plt.show()    
```

### 軸の設定をそろえる
- subplots()関数を使うと自動で軸の設定がそろえられる
- sharex,shareyオプションをTrueで軸範囲と目盛りが統一されている

```python
x1 = np.arange(0, 15, 0.2)
x2 = np.arange(0, 20, 0.2)

y1 = np.sin(x1)
y2 = np.sin(x2) * 2
y3 = np.cos(x2) * 2
y4 = np.cos(x1)

# フィギュアオブジェクトの生成
fig, axes = plt.subplots(2, 2, figsize=(3, 3), sharex=True, sharey=True)

# 描画
axes[0,0].plot(x1, y1)
axes[0,1].plot(x2, y2)
axes[1,0].plot(x2, y3)
axes[1,1].plot(x1, y4)

plt.show()
```

### 対数軸を設定する
- Axes.set_xscale(),Axes.set_yscale()に文字列logを渡す

```python
x1 = np.linspace(0, 100, 100)
x2 = np.linspace(0, 1000000, 100)
y1 = np.linspace(0, 1000000, 100)
y2 = np.linspace(0, 100, 100)

fig = plt.figure(figsize=(8,2))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.plot(x1, y1)
ax2.plot(x2, y2)

# 軸を対数に設定
ax1.set_yscale('log')
ax2.set_xscale('log')

ax1.set_title('Y軸が対数軸')
ax2.set_title('X軸が対数軸')

plt.show()
```

### 複数軸を設定する
- 2つの軸範囲を設定する場合はAxes.twinx()、Axes.twiny()メソッドを使用

```python
x1 = np.linspace(0, 10, 100)
x2 = np.linspace(0, 100, 100)
y1 = np.linspace(0, 100, 100)
y2 = np.linspace(0, 10, 100)

fig = plt.figure(figsize=(9,3))


ax1_1 = fig.add_subplot(121)
ax2_1 = fig.add_subplot(122)

ax1_1.plot(x1, y1)
ax1_2 = ax1_1.twinx()
ax1_2.plot(x1, y2)

ax2_1.plot(x1, y1)
ax2_2 = ax2_1.twiny()
ax2_2.plot(x2, y1)

ax1_1.set_title('2つのY軸', y=0.8)
ax2_1.set_title('2つのX軸', y=0.8)

plt.show()
```

## 軸ラベルを設定する
- 軸ラベルを設定するにはAxes.set_xlabel(),Axes.set_ylabel()

```python
x = np.arange(0, 15, 0.2)
y = np.sin(x)

fig = plt.figure(figsize=(4,2))
ax = fig.add_subplot(111)

ax.plot(x,y)

ax.set_xlim([0, 15])
ax.set_ylim([-1, 1])

# 軸ラベルの設定
ax.set_xlabel('x', fontsize=20, color='red')
ax.set_ylabel('y', {'fontsize': 20, 'color': 'green'})

fig.subplots_adjust(bottom=0.3, left=0.2)

plt.show()
```

### 主目盛りと主目盛りラベルを設定する
- 主目盛り: set_xticks,set_yticks ;任意の要素数のリスト(数値か真偽値)
- 主目盛りラベル: set_xticklabels, set_yticklabels: 主目盛りと同じ要素数のリスト(数値、真偽値、テキスト)

```python
fig = plt.figure(figsize=(4,2))
ax = fig.add_subplot(111)

ax.plot(x,y)

# 軸の範囲の設定
ax.set_xlim([0, 15])
ax.set_ylim([-1, 1])

# 主目盛りの設定
ax.set_xticks([0, 5, 10, 15])
ax.set_yticks([-1, 0, 1])

# 主目盛りラベルの設定
ax.set_xticklabels(['zero', 'five', 10, 15])
ax.set_yticklabels([])

fig.subplots_adjust(bottom=0.3, left=0.2)

plt.show()
```

### 目盛りの書式を設定する
- 目盛りの書式はAxes.tick_params()メソッドで設定する

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(111)

# 補助目盛りの描画
ax.minorticks_on()

# X軸の主目盛りを設定
ax.tick_params(axis='x',
              which='major',
              direction='inout',
              length=10,
              labelcolor='green')

# Y軸の補助目盛りを設定
ax.tick_params(axis='y',
              which='minor',
              direction='out',
              length=5,
              color='red')

# Y軸の主目盛りを設定
ax.tick_params(axis='y',
              which='major',
              direction='inout',
              length=10,
              color='blue')

plt.show()
```

### 目盛り線を設定する
- 目盛り線はAxes.grid()メソッドを用いる

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(111)

# 補助目盛りの描画
ax.minorticks_on()

# 目盛り線の描画
ax.grid(which='major', axis='y', color='blue', linestyle='-', linewidth=1)
ax.grid(which='minor', axis='y', color='grey', linestyle='--', linewidth=1)

plt.show()
```

### 5-7 凡例のタイトル
- 簡易な方法はデータ描画の際に引数labelに文字列を指定し、Axes.legend()で描画する

```python
x = np.arange(0, 20, 0.2)
y1 = np.sin(x)
y2 = np.cos(x)

fig =plt.figure()
ax = fig.add_subplot(111)

ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')

# 凡例の表示
ax.legend()

plt.show()
```

### 凡例を配置する
- 凡例を外側に配置したい場合はlocとbbox_to_anchorを用いる
- locは0～10で場所が指定出来る
- bbox_to_anchorは左下角の座標を(0,0)、右上角を(1,1)とした場合の相対座標

```python
fig = plt.figure()

ax = fig.add_subplot(111)

ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')

# 凡例表示・位置調整
ax.legend(loc='upper left', bbox_to_anchor=(1,1))

# サブプロットの位置の調整
fig.subplots_adjust(left=0.2, bottom=0.2, right=0.75, top=0.8)

plt.show()
```

### 凡例とタイトルの書式を設定する

```python
fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')

# フォントの書式辞書
fdict = {'family': 'serif',
        'style': 'italic',
        'weight': 'bold',
        'size': 12}

# テキストボックスの書式設定
bdict = {'facecolor': 'g',
        'edgecolor': 'None',
        'alpha': 0.5,
        'boxstyle': 'rarrow',
        'linewidth': 2}

ax.legend(loc='upper left', bbox_to_anchor=(1.02, 1), borderaxespad=0, prop=fdict)

# サブプロットタイトルの設定
ax.set_title('sin and cos', fontdict=fdict, bbox=bdict, color='white')

# フィギュアタイトルの設定
fig.suptitle('三角関数', y=0.95, size=16, color='grey', fontweight='bold')

fig.subplots_adjust(left=0.2, bottom=0.2, right=0.75, top=0.8)

plt.show()
```

## 5-9 ファイル出力

### ファイルを出力する
- ファイルの出力はsavefig()関数を使う
- 保存形式はformatオプションでPNG,PDF,PS,EPS,SVGを指定出来る

```python
# ファイルの出力
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)

dat = [1, 3, 2]

ax.plot(dat)

fname= 'save_fig.svg'
plt.savefig(fname)

plt.show()
```

### 出力するフィギュアオブジェクトの書式を設定する
1. ファイル出力オプション
- dpi : 解像度
- facecolor: 塗りつぶし色
- edgecolor: 枠線の色
- format: ファイルの保存形式
- transparent: 真偽値
- bbox_inches: 出力する図の大きさ、もしくはtight
- pad_inches: 余白幅
- frameon: 枠線を描画するか否か(真偽値)

```python
fig = plt.figure()
ax = fig.add_subplot(111)

dat = [1, 3, 2]

ax.plot(dat)

fname= 'save_fig_prop_setting.svg'

plt.savefig(fname, dpi=300, facecolor='grey',
          edgecolor='k', linewidht=10,
          frameon=True, tihgt_layout=False)

plt.show()
```

### 出力時にサイズを調整する
- 出力時の調整はsubplots_adjust()関数を使う

```python
x = np.arange(0, 15, 0.2)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure(figsize=(4,3), edgecolor='b', linewidth=5, frameon=True)

ax = fig.add_subplot(111)

ax.plot(x, y, label='sin')
ax.plot(x, z, label='cos')

ax.set_xlabel('x value')
ax.set_ylabel('y value')

ax.legend(loc='upper left', bbox_to_anchor=(1,1))

# サブプロットのサイズの調整
fig.subplots_adjust(left=0.2, bottom=0.2, right=0.75, top=0.9)

fname= 'save_fig_adjust.svg'
plt.savefig(fname, dip= 300, edgecolor='b', linewidth=2, frameon=True)

plt.show()
```
