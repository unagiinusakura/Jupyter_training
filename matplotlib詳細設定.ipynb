# 5.3 グラフの軸と目盛りの設定

```python
fig, axes = plt.subplots(2, 2)

axes[0, 0].plot(x1, y1)
axes[0, 1].plot(x2, y2)
axes[1, 0].plot(x3, y3)
axes[1, 1].plot(x4, y4)

axes[0, 0].set_xlim(0, 10)
axes[0, 1].set_xlim(10, 20)
axes[1, 0].set_xlim(20, 10)
axes[1, 1].set_xlim(10, 0)

axes[0, 0].set_ylim(0, 1)
axes[0, 1].set_ylim(0, -1)
axes[1, 0].set_ylim(-1, 0)
axes[1, 1].set_ylim(1, 0)

fig.subplots_adjust(hspace=0.3, wspace=0.3)
```

### 軸の形式を設定する

### 複数軸を設定する

```python
p = np.linspace(-10, 10, 20)
q = np.linspace(0, 10, 20)
r = np.linspace(-10, 10, 50)
s = [(np.tanh(ss /2 ) +1) /2 for ss in r]

fig, axes = plt.subplots(1, 2, figsize=(13, 4))

axes[0].scatter(p, q, color='blue', label='axes[0]')

ax2 = axes[0].twinx()
ax2.plot(r, s, color='blue', label='ax2')

axes[1].scatter(q, p, color='red', label='axes[1]')

ax3 = axes[1].twiny()
ax3.plot(s, r, color='red', label='ax3')

axes[0].set_title('2つのY軸')
axes[1].set_title('2つのX軸')

fig.legend();
```

```python
fig, ax1 = plt.subplots(1,1)

ax1.plot(p, q, color='blue', label='ax1')
ax2 = ax1.twinx()
ax2.plot(r, s, color='red', label='ax2')

ax1.set_ylabel('ax1', labelpad=-3, fontsize=20, color='blue')
ax2.set_ylabel('ax2', {'fontsize':20, 'color': 'red'})

fig.legend(bbox_to_anchor=(0.85, 0.25));
```

```python
fig, ax1 = plt.subplots(1, 1)

ax1.plot(p, q, color='blue', label='ax1')
ax2 = ax1.twinx()
ax2.plot(r, s, color='red', label='ax2')

# 主目盛りの設定
ax1.set_xticks([-10, 0, 10])
ax1.set_yticks(np.linspace(0, 10 ,3))
ax2.set_yticks(np.linspace(0, 1, 3))

# 主目盛りラベルの設定
ax1.set_yticklabels(['zero', 'five', 'ten'])
ax1.set_xticklabels([])

fig.legend(bbox_to_anchor=(0.85, 0.25));
```

### 目盛りの書式を設定する

```python
p = np.linspace(-10, 10, 20)
q = np.linspace(0, 10, 20)
r = np.linspace(-10, 10, 50)
s = [(np.tanh(ss /2 ) +1) /2 for ss in r]

fig, ax1 = plt.subplots(1, 1)

ax1.plot(p, q, color='blue', label='ax1')
ax2 = ax1.twinx()
ax2.plot(r, s, color='red', label='ax2')

# X軸の主目盛りを設定
ax1.set_xticks([-10, 0, 10])

# 補助目盛りの描画
ax1.minorticks_on()

# 目盛り線の描画
ax1.grid()

# Y軸(左軸)の主目盛りを設定
ax1.tick_params(axis='y',
                which='major',
                direction='inout',
                length = 10,
                width =3,
                color='blue',
                labelcolor='blue',
                labelrotation=90)

# Y軸(右軸)の主目盛りを設定
ax2.tick_params(axis='y',
                which='major',
                direction='inout',
                length = 10,
                width =3,
                color='red',
                labelcolor='red',
                labelrotation=-90)

# 左側の補助目盛りを設定
ax1.tick_params(axis='y',
                which='minor',
                direction='inout',
                length = 10,                
                color='blue',
                bottom='on',
                grid_color='red',
                grid_linewidth=1,
                grid_linestyle='-')

# X側の主目盛りを設定
ax1.tick_params(axis='x',
                which='major',
                direction='inout',
                length = 10,
                width=3,
                right='on',
                grid_linewidth=0
               )

# X側の補助盛りを設定
ax1.tick_params(axis='x',
                which='mainor',
                direction='inout',
                length = 5,
                right='on')

fig.legend(bbox_to_anchor=(0.8, 0.3));                
```

# 5.5 線の書式設定

```python
x = np.array([0,1])
y = [0, 1]

fig, ax = plt.subplots(1, 1)

# 線の太さと不透明度を設定して描画
ax.plot(x, y, linewidth=1, color='b', label='linewidth=1(alpha=1.0)')
ax.plot(x + 0.1, y, linewidth=1, color='b',alpha=0.5, label='linewidth=1(alpha=0.5)')

ax.plot(x + 0.3, y, linewidth=10, color='b', label='linewidth=10(alpha=1.0)')
ax.plot(x + 0.4, y, linewidth=10, color='b', alpha=0.5, label='linewidth=10(alpha=0.5)')

ax.legend(loc='upper left', bbox_to_anchor=(1,1));
```

#### 線の種類を設定する

```python
s = np.array([0, 1])
t = np.array([14, 14])

fig ,ax = plt.subplots()

# 線種を設定して描画
ax.plot(s, t, linewidth=5, color='blue', linestyle='solid', label='solid')
ax.plot(s, t-1, linewidth=5, color='skyblue', linestyle='-', label='-')

ax.plot(s, t-3, linewidth=5, color='blue', linestyle='dashed', label='dashed')
ax.plot(s, t-4, linewidth=5, color='skyblue', linestyle='--', label='--')

ax.plot(s, t-6, linewidth=5, color='blue', linestyle='dashdot', label='dashbot')
ax.plot(s, t-7, linewidth=5, color='skyblue', linestyle='-.', label='-.')

ax.plot(s, t-9, linewidth=5, color='blue', linestyle='dotted', label='dotted')
ax.plot(s, t-10, linewidth=5, color='skyblue', linestyle=':', label=':')

ax.plot(s, t-12, linewidth=5, color='orange', linestyle=(0, (5,2)), label='(0, (5,2))')
ax.plot(s, t-13, linewidth=5, color='orange', linestyle=(4, (5,2)), label='(0, (5,2))')

ax.set_ylim(0, 15)
ax.legend(loc='upper left', bbox_to_anchor=(1,1));
```

### 線の先端の形状を設定する

```python
x = [0, 1]
y = np.array([0.4, 0.4])

fig, ax = plt.subplots()

# 線の先端形状を設定して描画
ax.plot(x, y, linewidth=15, solid_capstyle='butt', label='butt')
ax.plot(x, y + 0.2, linewidth=15, solid_capstyle='round', label='round')
ax.plot(x, y + 0.4, linewidth=15, solid_capstyle='projecting', label='projecting')

ax.set_ylim([0, 1])
ax.legend(loc='lower right');
```

### 破線・点線の先端の形状を設定する

```python
fig, ax = plt.subplots()

# 破線・点線の先端形状を設定して描画
ax.plot(x, y, linewidth=15, linestyle='-.', dash_capstyle='butt', label='butt')
ax.plot(x, y+0.2, linewidth=15, linestyle='-.', dash_capstyle='round', label='round')
ax.plot(x, y+0.4, linewidth=15, linestyle='-.', dash_capstyle='projecting', label='projecting')

ax.set_ylim([0, 1])
ax.legend();
```

### 線の結合点の形状を設定する

```python
p = np.linspace(0, 4, 4)
q = [0, 1, 0, 1]

fig, ax = plt.subplots(figsize=(8, 3))

# 線の結合点の形状を設定して描画
ax.plot(p, q, linewidth=10, solid_joinstyle='miter', label='miter')
ax.plot(p + 0.5, q, linewidth=10, solid_joinstyle='round', label='round')
ax.plot(p + 1.0, q, linewidth=10, solid_joinstyle='bevel', label='bebel')
ax.set_ylim(-0.1, 1.1)
ax.legend(loc='lower right');
```

### 破線・点線の結合点の形状を設定する

```python
fig, ax = plt.subplots(figsize=(8, 3))

# 破線の結合形状を変えて描画
ax.plot(p, q, linewidth=10, linestyle='--', dash_joinstyle='miter', label='miter')
ax.plot(p + 0.5, q, linewidth=10, linestyle='--', dash_joinstyle='round', label='round')
ax.plot(p + 1.0, q, linewidth=10, linestyle='--', dash_joinstyle='bevel', label='bebel')

ax.set_ylim(-0.1, 1.1)
ax.legend(loc='lower right');
```

# 5.6 色の設定

### グラフ要素の色を設定する

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.arange(0, 6, 0.1)
y = np.sin(x)
z = np.cos(x)

fig, ax = plt.subplots(1, 1)

# 描画
ax.scatter(x, y, label='sin', color='lightblue', edgecolor='blue')
ax.scatter(x, z, label='cos', color='lightblue', edgecolor='blue', alpha=0.3)

ax.legend(loc='lower left');
```

```python
fig, ax = plt.subplots(1,1)

v = np.array([0, 1])

# 各種方法による色指定で描画
ax.plot(v, v, color='k', label='color string', linewidth=8)
ax.plot(v + 1, v, color='0.5', label='Gray scale', linewidth=8)
ax.plot(v + 2, v, color=(0, 0, 0.7, 1), label='RGBA tuple', linewidth=8)
ax.plot(v + 3, v, color='#66b3e9', label='RGBA 16', linewidth=8)
ax.plot(v + 4, v, color='DarkGreen', label='HTML color', linewidth=8)
ax.plot(v + 5, v, color='LimeGreen', label='X11/CSS', linewidth=8)
ax.plot(v + 6, v, color='C3', label='Regular expression', linewidth=8)
ax.plot(v + 7, v, color='tab:pink', label='Tableau', linewidth=8)

ax.legend(loc='upper left', bbox_to_anchor=(1, 1));
```

### カラーマップを利用する

```python
import matplotlib.cm as cm

fig, ax = plt.subplots(1, 1)

cname= 'viridis'
n = 10
cmap = cm.get_cmap(cname, n)

for i in range(n):
    ax.plot(x, y + n -i, color=cmap(i), label=f'{cname}({i})')
    
ax.legend(loc='upper left', bbox_to_anchor=(1, 1));    
```

```python
fig, ax = plt.subplots(1, 1)

cmap = cm.viridis

for i in np.linspace(0, 1, 10):
    ax.plot(x, y + 2 -i, color=cmap(i), label=f'viridis({i:.1f})')
    
ax.legend(loc='upper left', bbox_to_anchor=(1, 1));    
```

### 値に応じて段階的に色を変化させる

```python
fig, ax = plt.subplots(1, 1)

ax.scatter(x, y, c=y, cmap=cm.Blues);
```

### 個別に色を指定する

```python
fig, ax = plt.subplots(1, 1)

x = [0, 1]
y = np.array([0.4, 0.4])
cmap = cm.get_cmap('viridis')

ax.plot(x, y, color=cmap(0.0), label='cmap(0.0)')
ax.plot(x, y-0.2, color=cmap(0.6), label='cmap(0.6)')
ax.set_ylim(0, 0.6)

ax.legend();
```
